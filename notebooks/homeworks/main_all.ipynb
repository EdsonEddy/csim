```python
"""
Read a file and return its path and content.
Process all .py files in a given directory and store their names and contents.
"""

import os


def read_file(file_path):
    try:
        with open(file_path, "r", encoding="utf-8") as file:
            content = file.read()
        return file_path, content
    except Exception as e:
        print(f"Error reading file {file_path}: {e}")
        return file_path, None


def process_files(target_directory):
    # Storage for file names and contents
    file_names = []
    file_contents = []

    # Check if the path is a valid directory
    if not os.path.isdir(target_directory):
        print(f"Error: The path '{target_directory}' is not a valid directory.")
        return file_names, file_contents

    for dirs in os.walk(target_directory):
        # Get the current directory path
        path = dirs[0]
        # Process the files in the directory
        for file in os.listdir(path):
            file_path = os.path.join(path, file)
            if os.path.isfile(file_path) and file.endswith(".py"):
                file_name, content = read_file(file_path)
                # Store the file name and content
                file_names.append(file_name)
                file_contents.append(content)

    return file_names, file_contents
```

```python
import os

# Get the current working directory and construct the path to the homework files
files_path = os.path.abspath(os.getcwd())
parent_path = os.path.dirname(files_path)
# Assuming the homework files are located in a directory named "datasets/INF-111-JT-Tarea 1-6876" relative to the current working directory
homework_directory = os.path.join(parent_path, "datasets", "INF-111-JT-Tarea 1-6876")
file_names, file_contents = process_files(homework_directory)
```
